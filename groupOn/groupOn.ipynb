```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
```

```python
driver = webdriver.Firefox(executable_path='geckodriver.exe')
driver.get("https://www.groupon.co.uk/#")
wait = WebDriverWait(driver, 10)
time.sleep(5)
driver.find_element_by_xpath('//*[@id="nothx"]').click()
time.sleep(2)

driver.find_element_by_xpath('/html/body/div[3]/div/div/div[2]/button').click()
search_input = wait.until(EC.presence_of_element_located((By.ID, "ls-search")))
search_input.clear()
search_input.send_keys('Writing Courses')
driver.find_element_by_id('ls-header-search-button').click()
time.sleep(5)
elems = driver.find_elements_by_xpath("//a[@href]")
pageNumber = driver.find_element_by_xpath("/html/body/main/div/div/div[3]/div[4]/div/div/div[2]").text.split("\n")[-1]
print(pageNumber)
links = list()
```

```python
for count in range(int(pageNumber)):
    print(count)
    for elem in elems:
        try:
            print(elem.get_attribute("href"))
            if "https://www.groupon.co.uk/deals/" in elem.get_attribute("href"):
                links.append(elem.get_attribute("href"))
                print(elem.get_attribute("href"))
        except:
            pass
    driver.find_element_by_xpath('//*[@id="next-page"]').click()
    time.sleep(5)
print(len(links))
```

```python
tempLinks = list()
tempLinks = links
for index in range(len(links)):
    for index2 in range(len(links)):
        if tempLinks[index] in tempLinks[index2] and tempLinks[index] != tempLinks[index2]:
            tempLinks[index2] = tempLinks[index]
        elif tempLinks[index] == tempLinks[index2]:
            pass
        else:
            pass

print(tempLinks)
print(len(tempLinks))
    
```

```python
for item in tempLinks:
    print(item)
print(len(tempLinks))
```

```python
links = list(set(tempLinks))
print(len(links))
```

```python
import pickle
with open("links.pickle", "wb") as f:
    pickle.dump(links, f)
with open("links.pickle", "rb") as f:
    links = pickle.load(f)
```

```python
from bs4 import BeautifulSoup
import requests

class soup:
    def __init__(self, pageLink, parser):
        self.link = pageLink
        self.parser = parser

    def pageResponse(self):
        page_response = requests.get(self.link, timeout=5)
        return BeautifulSoup(page_response.content, self.parser)
    
```

```python
driver = webdriver.Firefox(executable_path='geckodriver.exe')
data = list()
for link in links:
    driver.get(link)
    wait = WebDriverWait(driver, 10)
    time.sleep(5)
    try:
        driver.find_element_by_xpath('//*[@id="nothx"]').click()
    except:
        pass
    provider = driver.find_element_by_xpath('//*[@id="deal-title"]').text
    
    title = driver.find_element_by_xpath('//*[@id="inline-buy"]/fieldset/ul/li/div/label/div/h3').text
    
    price = driver.find_element_by_xpath('//*[@id="inline-buy"]/fieldset/ul/li/div/label/div/div/div[2]/div[1]/div[2]/div').text

    purchased = driver.find_element_by_xpath('//*[@id="inline-buy"]/fieldset/ul/li/div/label/div/div/div[1]/div').text
    purchased = purchased.replace(" bought", "")
    tmpData = {
                    "courseLink": link,
                    "courseName": title,
                    "coursePrice": price,
                    "courseProvider": provider,
                    "Purchase": purchased
                }
    print(tmpData)
    data.append(tmpData.copy())
    
print(data)
```

```python
print(len(data))

keyword = ("write", "writer", "writers", "written", "writing")
wData = list()
for item in data:
    for key in keyword:
        if key in item['courseName']:
            wData.append(item.copy())
            print(item)
try:
    with open("wData.pickle", "wb") as f:
        pickle.dump(wData, f)
except Exception as e:
        print(e)
```
